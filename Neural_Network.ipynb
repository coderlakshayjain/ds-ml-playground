```python
import matplotlib.pyplot as plt

def draw_neural_net(ax, layer_sizes):
    """
    Draws a neural network with given layer sizes.
    :param ax: Matplotlib Axes object
    :param layer_sizes: List of integers representing the number of neurons in each layer
    """
    left = 0.1  # position of the leftmost layer
    right = 0.9  # position of the rightmost layer
    bottom = 0.5  # bottom-most position
    top = 0.9  # top-most position
    v_spacing = (top - bottom) / float(max(layer_sizes))  # vertical spacing between neurons
    h_spacing = (right - left) / float(len(layer_sizes) - 1)  # horizontal spacing between layers

    # Draw neurons
    for n, layer_size in enumerate(layer_sizes):
        layer_top = v_spacing * (layer_size - 1) / 2 + bottom
        for m in range(layer_size):
            circle = plt.Circle((n * h_spacing + left, layer_top - m * v_spacing), v_spacing / 4,
                                color='w', ec='k', zorder=4)
            ax.add_artist(circle)

    # Draw edges (connections between neurons)
    for n, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
        layer_top_a = v_spacing * (layer_size_a - 1) / 2 + bottom
        layer_top_b = v_spacing * (layer_size_b - 1) / 2 + bottom
        for m in range(layer_size_a):
            for o in range(layer_size_b):
                line = plt.Line2D([n * h_spacing + left, (n + 1) * h_spacing + left],
                                  [layer_top_a - m * v_spacing, layer_top_b - o * v_spacing], c='k')
                ax.add_artist(line)

# Parameters for the neural network
layer_sizes = [3, 5, 2]  # 3 neurons in input, 5 in hidden, and 2 in output layer

# Plotting the neural network
fig = plt.figure(figsize=(6, 6))
ax = fig.gca()
ax.axis('off')
draw_neural_net(ax, layer_sizes)
plt.show()
```
